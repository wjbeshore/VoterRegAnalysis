# Voter Registration comparison 2016 compared 2020

We compare the totals voter registration in Arizona, California, Colorado, Delaware, DC, Florida, Georgia, Illinois, Maryland, North Carolina, Texas, and Virginia. We compare the total percent changes month to month in these states during primaries.

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import plotly


file = pd.read_csv("new-voter-registrations.csv")
print(file.head(10))
```

```python
stateArray = file.Jurisdiction.unique()
filter2016 = file['Year'] == 2016
filter2020 = file['Year'] == 2020
print(file[(file['Jurisdiction'] == stateArray[2]) & filter2016])
```

```python


for item in stateArray:
    temp_data16 = file[(file['Jurisdiction'] == item) & filter2016]
    temp_data20 = file[(file['Jurisdiction'] == item) & filter2020]
    plt.plot(temp_data16['Month'], temp_data16["New registered voters"], label="2016")
    plt.plot(temp_data20['Month'], temp_data20["New registered voters"], label="2020")
    plt.title(item)
    plt.legend()
    plt.show()

```
