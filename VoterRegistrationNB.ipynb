```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import plotly


file = pd.read_csv("new-voter-registrations.csv")
print(file.head(10))
```

```python
stateArray = file.Jurisdiction.unique()
filter2016 = file['Year'] == 2016
filter2020 = file['Year'] == 2020
print(file[(file['Jurisdiction'] == stateArray[2]) & filter2016])
```

```python
sns.set()
sns.set_style("white")
sns.set_palette("gray")
sns.despine(left=True, bottom=True)
for item in stateArray:
    temp_data16 = file[(file['Jurisdiction'] == item) & filter2016]["New registered voters"]
    temp_data20 = file[(file['Jurisdiction'] == item) & filter2020]["New registered voters"]
    temp_data16.plot.line(grid = False).set_title("New registered voters " + item)
    temp_data20.plot.line(grid = False).set_title("New registered voters " + item)
    plt.show()

    

```

```python
import plotly.express as px

fig = px.line(x=["Jan","Feb","Mar", "Apr"], y=file[(file['Jurisdiction'] == "Colorado") & filter2016]["New registered voters"], title="sample figure")
fig = px.line(x=["Jan","Feb","Mar", "Apr"], y=file[(file['Jurisdiction'] == "Colorado") & filter2020]["New registered voters"], title="sample figure")
print(fig)
fig.show()
```
